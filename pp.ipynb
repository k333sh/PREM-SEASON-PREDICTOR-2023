```python
#we need to get numerica data types
matches["date"] = pd.to_datetime(matches["date"])
```

```python
matches["VenueCode"] = matches["venue"].astype("category").cat.codes #0 away , 1 home
```

```python
matches["oppCode"] = matches["opponent"].astype("category").cat.codes
```

```python
matches["hour"] = matches["time"].str.replace(":.+", "", regex=True).astype("int")
```

```python
matches["dayCode"] = matches["date"].dt.day_of_week
```

```python
matches["target"] = (matches["result"]== "W").astype("int")
```

```python
#random forest cause it picks up non linear relationships a regular linear model would not
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators= 80, min_samples_split= 8 , random_state = 1 )
```

```python
train = matches[matches["date"] < '2022-01-01']
```

```python
test = matches[matches["date"] > '2022-01-01']
```

```python
predictors = ["VenueCode","oppCode","hour","dayCode"]
```

```python
rf.fit(train[predictors],train["target"])
```

```python
preds = rf.predict(test[predictors])
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc = accuracy_score(test["target"], preds)
acc
```

```python
groupedMatches = matches.groupby("team")
```

```python
group = groupedMatches.get_group("Arsenal")
group
```

```python
def rolling_averages(group, cols,newCols):
    group = group.sort_values("date")
    rollingStats = group[cols].rolling(3,closed="left").mean()
    group[newCols] = rollingStats
    group = group.dropna(subset=newCols) # prevent pandas from placing na values without 3 week minimum played
    return group
```

```python
cols = ["gf","ga","sh","sot","dist","fk","pk","pkatt"]
newCols = [f"{c}_rolling" for c in cols]
```

```python
newCols
```

```python
rolling_averages(group, cols, newCols)
```

```python
matches_rolling = matches.groupby("team").apply(lambda x: rolling_averages(x, cols, newCols), include_groups=False)
matches_rolling
```

```python
matches_rolling.index = range(matches_rolling.shape[0])
```

```python
def make_predictions(data, predictors):
    train = data[data["date"] < '2022-01-01']
    test = data[data["date"] > '2022-01-01']
    rf.fit(train[predictors],train["target"])
    preds = rf.predict(test[predictors])
    combined = pd.DataFrame(dict(actual=test["target"], predicted=preds) , index = test.index)
    precision = precision_score(test["target"], preds)
    return combined , precision
```

```python
combined, precision = make_predictions(matches_rolling, predictors + newCols)
```
